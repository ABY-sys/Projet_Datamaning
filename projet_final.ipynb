# ANALYSE DE L'ECOSYSTÈME ECONOMIQUE DE L'AFRIQUE






## NOTRE JEU DE DONNÉES

Les données représentent les indicateurs de développement mondial mesurées de 1960 à 2022, extraites de la base de données de la Banque mondiale. Il comprend des données macroéconomiques, sociales, politiques et environnementales de tous les pays et régions sur lesquels la banque mondiale dispose de données. 
### Colonnes incluses
1. **pays** : Le pays.
2. **date** : Date de la mesure. 
3. **agricultural_land%** : Terres agricoles en % de la superficie du pays.
4. **forest_land%** : Superficie forestière en tant que % de la superficie terrestre du pays.
5. **land_area** : Superficie terrestre, mesurée en km^2.
6. **avg_precipitation** : Précipitations moyennes en profondeur, mesurées en mm par an.
7. **trade_in_services%** : Commerce des services en % du PIB.
8. **control_of_corruption_estimate** : Index qui estime le contrôle de la corruption.
9. **control_of_corruption_std** : Erreur standard de l'estimation du contrôle de la corruption.
10. **access_to_electricity%** : Pourcentage de la population qui a accès à l'électricité.
11. **renewable_energy_consumption%** : Consommation d'énergie renouvelable en % de la consommation finale totale d'énergie.
12. **electric_power_consumption** : Consommation d'énergie électrique, mesurée en kWh par habitant.
13. **CO2_emissions** : Émissions de CO2, mesurées en kt.
14. **other_greenhouse_emisions** : Émissions totales de gaz à effet de serre, mesurées en kt d'équivalent CO2.
15. **population_density** : Densité de population, mesurée en personnes par km^2 de superficie terrestre.
16. **inflation_annual%** : Inflation, prix à la consommation, en pourcentage annuel.
17. **real_interest_rate** : Taux d'intérêt réel (%).
18. **risk_premium_on_lending** : Prime de risque sur le prêt (taux de prêt moins taux de bon du Trésor, %).
19. **research_and_development_expenditure%** : Dépenses de recherche et de développement, en pourcentage du PIB.
20. **central_goverment_debt%** : Dette du gouvernement central, total, en % du PIB.
21. **tax_revenue%** : Recettes fiscales en % du PIB.
22. **dépense %** : Dépense en % du PIB.
23. **government_effectiveness_estimate** : Index qui estime l'efficacité du gouvernement.
24. **government_effectiveness_std** : Erreur standard de l'estimation de l'efficacité du gouvernement.
25. **human_capital_index** : Indice du capital humain (HCI) (échelle 0-1).
26. **doing_business** : Score de facilité de faire des affaires (0 = performance la plus basse à 100 = meilleure performance).
27. **time_to_get_operation_license** : Jours nécessaires pour obtenir une licence d'exploitation.
28. **statistical_performance_indicators** : Indicateurs de performance statistique (SPI) : Score global (échelle 0-100).
29. **individuals_using_internet%** : Pourcentage de la population utilisant Internet.
30. **logistic_performance_index** : Indice de performance logistique : Global (1=faible à 5=élevé).
31. **military_expenditure%** : Dépenses militaires en % du PIB.
32. **GDP_current_US** : PIB (US$ actuel).
33. **political_stability_estimate** : Index qui estime la stabilité politique et l'absence de violence/terrorisme.
34. **political_stability_std** : Erreur standard de l'estimation de la stabilité politique et de l'absence de violence/terrorisme.
1. **rule_of_law_estimate** : Index qui fait une estimation de l'État de droit.
2. **rule_of_law_std** : Erreur standard de l'estimation de l'État de droit.
3. **regulatory_quality_estimate** : Indice qui fait une estimation de la qualité réglementaire.
4. **regulatory_quality_std** : Erreur standard de l'estimation de la qualité réglementaire.
5. **government_expenditure_on_education%** : Dépenses publiques en matière d'éducation, total, en % du PIB.
6. **government_health_expenditure%** : Dépenses nationales de santé des administrations publiques en % du PIB.
7. **multidimensional_poverty_headcount_ratio%** : Ratio multidimensionnel de la pauvreté (% de la population totale).
8. **gini_index** : indice Gini.
9. **taux_de_naissance** : Taux de natalité, brut (pour 1 000 personnes).
10. **death_rate** : Taux de mortalité, brut (pour 1 000 personnes).
11. **life_expectancy_at_birth** : Espérance de vie à la naissance, total (années).
12. **population** : Population totale.
13. **population_rural** : population rurale.
14. **voice_and_accountability_estimate** : Index qui fait une estimation de la voix et de la responsabilité.
15. **voice_and_accountability_std** : Erreur standard de l'estimation de la voix et de la responsabilité.
16. **intentional_homicides** : homicides intentionnels (pour 100 000 personnes).








```python
import pandas as pd
```

```python
import pycountry

# Obtenir la liste des codes ISO alpha-2 des pays africains
codes = [country.alpha_3 for country in pycountry.countries if hasattr(country, 'region') and country.region == 'AF']

# Obtenir la liste des noms des pays africains correspondants
pays = [country.name for country in pycountry.countries if country.alpha_3 in codes]
# Filtrer les données pour ne garder que les pays africains
df_pays_africains = df[df['country'].isin(pays)]

df['Year'] = df['date'].str.slice(0, 4)
df['Year'] = pd.to_datetime(df['Year'])

#print(df['Year'].dtype)
# Afficher les valeurs nulles dans le DataFrame
print(df.isnull().sum())
df.shape

```

```python
import matplotlib.pyplot as plt

# Calculer le pourcentage de valeurs nulles dans chaque colonne
null_percentage = (df.isnull().sum() / len(df)) * 100

# Tracer un graphique à barres pour visualiser le pourcentage de valeurs nulles dans chaque colonne
plt.figure(figsize=(10, 6))
null_percentage.plot(kind='bar')
plt.title('Pourcentage de valeurs nulles par colonne')
plt.xlabel('Colonnes')
plt.ylabel('Pourcentage de valeurs nulles')
plt.xticks(rotation=90)
plt.show()
```

```python


# Calculer le pourcentage de valeurs nulles par année
null_percentage_by_year = df.groupby('Year').apply(lambda x: (x.isnull().sum() / len(x)) * 100)

# Tracer un graphique pour visualiser le pourcentage de valeurs nulles par année
plt.figure(figsize=(12, 8))
null_percentage_by_year.plot(kind='line', marker='o')
plt.title('Pourcentage de valeurs nulles par année')
plt.xlabel('Année')
plt.ylabel('Pourcentage de valeurs nulles')
plt.xticks(rotation=45)
plt.grid(True)
#plt.legend(loc='upper right')
plt.tight_layout()
plt.show()
```
